```python
def download_and_extract_from_drive(file_id, extract_to):
    url = f"https://drive.google.com/uc?id={file_id}"

    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()  # Raise an exception for bad status codes (4xx or 5xx)

        z = zipfile.ZipFile(io.BytesIO(response.content))
        z.extractall(extract_to)
        print(f"File successfully extracted to: {extract_to}")

    except requests.exceptions.RequestException as e:
        print(f"Error downloading file: {e}")
    except zipfile.BadZipFile as e:
        print(f"Error: Invalid zip file: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

#https://drive.google.com/file/d/1IKnPFnFzsDb4kSI3RPzL_0GOcq0ebo4a/view?usp=drive_link
#https://drive.google.com/file/d/1IKnPFnFzsDb4kSI3RPzL_0GOcq0ebo4a/view?usp=sharing

file_id = "1IKnPFnFzsDb4kSI3RPzL_0GOcq0ebo4a"  # Replace with your file ID
extract_to = "/content/Results"  # Replace with your desired extraction path

os.makedirs(extract_to, exist_ok=True)

download_and_extract_from_drive(file_id, extract_to)

def read_cont(file_path):
    with open(file_path, 'r') as file:
        file_content = file.read()

        global_accuracies = ast.literal_eval(file_content.split('=')[1].strip())
    return global_accuracies
```

# Graph Code

```python
def antiflipper_graph(dataset,data_distribution,type,n=25):
  listt = []

  if type == "Trust":
    listt = trust_data(dataset,data_distribution)
  if type == "Accuracy":
    listt = accuracy_data(dataset,data_distribution)


  list_n = listt[:n]  # Takes only first 25 values as input

  # List to store average trust of honest and malicious nodes per round
  bad_avg = []
  good_avg = []

  # List of nodes labelled as malicious by AntiFLipper
  detected_nodes = read_cont(str('/content/Results/FINAL RESULTS/'+data_distribution+'/'+dataset+'/'+ 'AntiFLipper'+'/0.4/'+"detected_nodes.txt"))

  # Compute average trust for each round in the first 25 rounds
  for t in list_n:
      bad_values = [v for k, v in t.items() if k in detected_nodes]
      good_values = [v for k, v in t.items() if k not in detected_nodes]

      bad_avg.append(np.mean(bad_values) if bad_values else 0)  # Avg trust of bad nodes
      good_avg.append(np.mean(good_values) if good_values else 0)  # Avg trust of good nodes

  # X-axis: first 25 rounds
  x = list(range(n))

  plt.figure(figsize=(8, 5))

  plt.plot(x, good_avg, marker='o', linestyle='-', markersize=5, linewidth=1.5, alpha=0.8, color='green', label=f'Avg. {type} of Honest Clients')
  plt.plot(x, bad_avg, marker='o', linestyle='--', markersize=6, linewidth=2, alpha=0.8, color='red', label=f'Avg. {type} of Malicious Clients')


  fs = 22
  plt.xlabel("Rounds (First 25)", fontsize=fs) #, fontweight='bold')
  plt.ylabel(f"Average {type} Value", fontsize=fs) # ,fontweight='bold')
  plt.xticks(fontsize=fs) #,fontweight='bold')
  plt.yticks(fontsize=fs) #,fontweight='bold')
  plt.legend(fontsize=16)
  plt.grid(True, linestyle='--', alpha=0.6)
  #plt.ylim(-0.05, 1.03)
  plt.tight_layout()
  filename = f"antiflipper_graph_{dataset}_{data_distribution}_{type}.pdf"
  plt.savefig(filename, format='pdf')
  plt.show()
```
